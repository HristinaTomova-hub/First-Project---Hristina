# Pet Growth Analysis

* This project analyzes how a pet's weight changes over time using Python.

* The goal is to visualize the growth of a pet and perform basic statistical analysis.

```python
import matplotlib.pyplot as plt

age = [1, 2, 3, 4, 5, 6]
weight = [2, 3.5, 5, 7, 9, 12]
```

```python
plt.plot(age, weight)
plt.scatter(age, weight)

plt.title("Pet Growth Over Time")
plt.xlabel("Age (months)")
plt.ylabel("Weight (kg)")

plt.grid(True)
plt.show()
```

## Analysis

* The graph shows that the pet's weight increases as it grows older.

* This simple model demonstrates how data visualization can help analyze trends in growth.

## Mean weight

* The mean weight is calculated as:

$$\bar{w} = \frac{w_1 + w_2 + \dots + w_n}{n}$$

```python
mean_weight = sum(weight) / len(weight)
mean_weight
```

The computed mean weight for the selected period is shown above.

```python
growth = []
for i in range(1, len(weight)):
    growth.append(weight[i] - weight[i-1])

growth
```

```python
plt.plot(age[1:], growth)
plt.scatter(age[1:], growth)
plt.title("Monthly Weight Gain")
plt.xlabel("Age (months)")
plt.ylabel("Weight gain (kg)")
plt.grid(True)
plt.show()
```

* Interpretation

The trend suggests increasing weight with age.
The histogram gives an idea of how the observed weights are distributed.

* Basic statistics

We compute basic statistics for the pet's weight data.

Mean (average) weight:

$$\bar{w}=\frac{w_1+w_2+\dots+w_n}{n}$$

```python
import numpy as np

mean_w = np.mean(weight)
min_w = np.min(weight)
max_w = np.max(weight)
std_w = np.std(weight)

mean_w, min_w, max_w, std_w
```

```python
plt.hist(weight, bins=5)
plt.title("Weight distribution (Histogram)")
plt.xlabel("Weight (kg)")
plt.ylabel("Count")
plt.grid(True)
plt.show()
```

* Interpretation (based on statistics)

- The mean weight summarizes the typical weight in this period.
- The standard deviation shows how much the weights vary.
- The histogram visualizes how the weights are distributed across the months.
